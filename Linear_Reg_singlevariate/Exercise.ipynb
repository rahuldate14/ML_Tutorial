```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df.head()
```

```python
df.rename(columns={"per capita income (US$)": "per_capita_income"}, inplace=True)
df.head()
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.scatter(df.year, df.per_capita_income, color="red", marker="+")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["year"]], df.per_capita_income)
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.title("Per Capita Income in Canada")
plt.scatter(df.year, df.per_capita_income, color="red", marker="+", label="data points")
plt.plot(df.year, reg.predict(df[["year"]]), color="blue", label="regression line")
plt.legend()
plt.show()
```
