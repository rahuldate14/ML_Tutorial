```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('HR_comma_sep.csv')
df.head()
```

```python
df.shape
```

```python
left = df[df.left == 1]
left.shape
```

```python
remained = df[df.left == 0]
remained.shape
```

```python
df.groupby('left').mean(numeric_only=True)
```

```python
pd.crosstab(df.salary, df.left).plot(kind='bar')
plt.xlabel('Salary')
plt.ylabel('Number of Employees')
plt.title('Salary vs Left')
plt.show()
```

```python
pd.crosstab(df.Department, df.left).plot(kind='bar')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.title('Department vs Left')
plt.show()
```

```python
subset = df[['satisfaction_level', 'average_montly_hours', 'promotion_last_5years', 'salary']]
subset.head()
```

```python
salary_dummies = pd.get_dummies(subset.salary, prefix='salary')
salary_dummies.head()
```

```python
df_final = pd.concat([subset, salary_dummies], axis='columns')
df_final.head()
```

```python
df_final.drop('salary', axis='columns', inplace=True)
df_final.head()
```

```python
x = df_final
x.head()
```

```python
y = df.left
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test, y_test)
```
